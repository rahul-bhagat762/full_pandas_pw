##merge and join

```python
df_1 = {'key1':(1,2,3,4),
    'key2':(5,6,7,8),
    'key3':(8,9,10,11)}
df_1 = pd.DataFrame(df_1)
```

```python
df_2 = {'key1':(45,87,98,97),
    'key4':(78,98,6,45),
    'key5':(98,65,54,99)}
df_2 = pd.DataFrame(df_2)
```

```python
df_1
```

```python
df_2
```

```python
##Merge: It will happen on column but join happen on indexes
pd.merge(df_1,df_2)
```

```python
##Merge
##left Join- All the element from left side and only take common elements from right side 
pd.merge(df_1,df_2,how= 'left')
```

```python
##Merge
##Right Join- All the element from right side and only take common elements from left side 
pd.merge(df_1,df_2,how= 'right')
```

```python
##Merge
##Outer Join- All the keys from both the data when this same column name is present and assign elements of second table down to the first table 
#and when this same name table is not preent then assign nan value
pd.merge(df_1,df_2,how='outer')
```

```python
pd.merge(df_1,df_2,how='cross') ##it is try to create all the possible combination
```

```python
pd.merge(df_1,df_2,how='left',left_on='key2',right_on='key4')
```
